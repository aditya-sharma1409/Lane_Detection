# SVR

Linear SVR Kernel

```python
#import the libraries
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.svm import SVR 

# Generating a cosh wave datapoints
X_data = np.sort(5 * np.random.rand(40, 1), axis=0) 
y_data = np.cosh(X_data).ravel()  #conerting the cosh data into a 1D array

# add noise to every 5th datapoint
y_data[::5] += 3 * (0.5 - np.random.rand(8)) 

# Create an SVR model with a linear kernel 
svr_model = SVR(kernel='linear') 

# Training the model 
svr_model.fit(X_data, y_data) 

# Predicting
y_pred_data = svr_model.predict(X_data) 

# Plot the predicted values against the true values 
plt.scatter(X_data, y_data, color='orange', label='Actual data') 
plt.plot(X_data, y_pred_data, color='blue', label='Predicted values') 
plt.legend() 
plt.show()
```

Polynomial SVR Kernel

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.svm import SVR 

# Generating a cosh wave datapoints
X_custom = np.sort(5 * np.random.rand(40, 1), axis=0) 
y_custom = np.cosh(X_custom).ravel() 


y_custom[::5] += 3 * (0.5 - np.random.rand(8)) 

# Create an SVR model with a polynomial kernel 
svr_custom = SVR(kernel='poly') 


svr_custom.fit(X_custom, y_custom) 

# Predicting
y_pred_custom = svr_custom.predict(X_custom) 


plt.scatter(X_custom, y_custom, color='purple', label='Actual data') 
plt.plot(X_custom, y_pred_custom, color='green', label='Predicted values') 
plt.legend() 
plt.show()
```

Radial Basis SVR Kernel

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.svm import SVR 

# Generating a cosh wave datapoints
X_custom = np.sort(5 * np.random.rand(40, 1), axis=0) 
y_custom = np.cosh(X_custom).ravel() 


y_custom[::5] += 3 * (0.5 - np.random.rand(8)) 

# Create an SVR model with a radial basis function (RBF) kernel 
svr_custom = SVR(kernel='rbf') 


svr_custom.fit(X_custom, y_custom) 

# Predicting
y_pred_custom = svr_custom.predict(X_custom) 


plt.scatter(X_custom, y_custom, color='red', label='Actual data') 
plt.plot(X_custom, y_pred_custom, color='magenta', label='Predicted values') 
plt.legend() 
plt.show()
```
